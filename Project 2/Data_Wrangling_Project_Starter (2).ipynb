### **Quality Issue 1: FILL IN**

```python
# FILL IN - Apply the cleaning strategy
# handling missing values

df1.dropna(subset=['Rating','Type'], inplace=True)


df1.dropna(subset=['Content Rating'], inplace=True)

#i will fill what is missing from the last two varibles with
# 'depndes on device' so that we avoid dropping

df1['Current Ver'] = df1['Current Ver'].fillna('Varies with device')
df1['Android Ver'] = df1['Android Ver'].fillna('Varies with device')
```

```python
#this to correct the dtypes of the set 

df1['Reviews']=pd.to_numeric(df1['Reviews'])

df1['Installs'] = df1['Installs'].str.replace(',', '').str.replace('+', '').astype(int)


df1['Price'] = df1['Price'].str.replace('$', '').astype(float)

df1['Last Updated']= pd.to_datetime(df1['Last Updated'])
```

```python
# FILL IN - Validate the cleaning was successful

df1.isnull().sum()
```

```python
#this to ensure the dtypes have changed correctly 

df1.info()
```
